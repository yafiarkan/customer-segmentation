## Data Source and Dictionary

Dataset diperoleh dari Kaggle (https://www.kaggle.com/arjunbhasin2013/ccdata). Dataset berisi aktivitas 9000 pelanggan aktif kartu kredit selama 6 bulan. Terdapat 18 kolom yang digunakan untuk menggambarkan informasi, aktivitas dan jenis pelanggan.


1. CUSTID : No identifikasi pemegang kartu kredit


2. BALANCE : Jumlah balance yang tersisa untuk melakukan pembelian


3. BALANCEFREQUENCY : Frekuensi perubahan balance, skor antara 0 dan 1(0=tidak sering, 1=sering)


4. PURCHASES : Jumlah pembelian


5. ONEOFFPURCHASES : Maksimum pembelian dalam suatu transaksi 


6. INSTALLMENTSPURCHASES : Jumlah pembelian saat installment


7. CASHADVANCE : Cash in advance given by the user

8. PURCHASESFREQUENCY : How frequently the Purchases are being made, score between 0 and 1 (1 = frequently purchased, 0 = not frequently purchased)

9. ONEOFFPURCHASESFREQUENCY : How frequently Purchases are happening in one-go (1 = frequently purchased, 0 = not frequently purchased)

10. PURCHASESINSTALLMENTSFREQUENCY : How frequently purchases in installments are being done (1 = frequently done, 0 = not frequently done)

11. CASHADVANCEFREQUENCY : How frequently the cash in advance being paid 

12. CASHADVANCETRX : Number of Transactions made with "Cash in Advanced"

13. PURCHASESTRX : Numbe of purchase transactions made

14. CREDITLIMIT : Limit of Credit Card for user

15. PAYMENTS : Amount of Payment done by user

16. MINIMUM_PAYMENTS : Minimum amount of payments made by user

17. PRCFULLPAYMENT : Percent of full payment paid by user

18. TENURE : Tenure of credit card service for user

# K-Means dan Mini batch K-means Clustering

Sebelum melakukan clustering, akan diselidiki terlebih dahulu berapa jumlah cluster terbaik dari dataset.

<h2> Metode Elbow K-Means </h2>
Metode elbow merupakan suatu metode yang digunakan untuk menghasilkan informasi dalam menentukan jumlah cluster terbaik.

```python
#Metode elbow untuk K-Means
with plt.style.context('seaborn'):
    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
        kmeans.fit(X)
        wcss.append(kmeans.inertia_)
    plt.figure(figsize=(12,8))
    plt.xticks(range(1,11),fontsize=11)
    plt.plot(range(1, 11), wcss)
    plt.title('Metode Elbow pada Dataset Pelanggan Kartu Kredit (k-Means)', fontsize=20)
plt.show()
```

Dari grafik di atas diperoleh jumlah kluster terbaik adalah 2, 4 atau 5.

<h2> Metode Elbow Mini Batch K-Means </h2>

```python
# Metode Elbow
from sklearn.cluster import MiniBatchKMeans

distortions = []
K = range(1,11)
for k in K:
    MBKMeanModel = MiniBatchKMeans(n_clusters=k, random_state=42)
    MBKMeanModel.fit(Xm)
    distortions.append(MBKMeanModel.inertia_)

plt.figure(figsize=(12,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('Metode Elbow pada Dataset Pelanggan Kartu Kredit(Mini Batch K-Means)', fontsize=20)
plt.show()
```

Dari grafik di atas diperoleh jumlah kluster terbaik adalah 2 atau 4.

<h2> Metode Silhouette K-means</h2>
Metode Silhouette
Silhouette Coefficient digunakan untuk melihat kualitas dan kekuatan cluster, seberapa baik suatu objek ditempatkan dalam suatu cluster.

```python
with plt.style.context('seaborn'):
    silhouette_scores = [] 
    for n_cluster in range(2, 11):
        silhouette_scores.append(   
            silhouette_score(X, KMeans(n_clusters = n_cluster).fit_predict(X))) 

    k = [2, 3, 4, 5, 6,7,8,9,10] 
    plt.figure(figsize=(10,8))
    plt.bar(k, silhouette_scores, color='red') 
    plt.title('Silhoutte Score Credit Card User (K-Means)',fontsize=20)
    plt.xticks(range(1,11),fontsize=11)
    plt.xlabel('Number of clusters', fontsize = 10) 
    plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show()    
```

Dari grafik di atas diperoleh jumlah kluster terbaik adalah 2, 4 atau 5.

<h2> Metode Silhouette Mini Batch K-means</h2>

```python
with plt.style.context('seaborn'):
    silhouette_scores = [] 
    for n_cluster in range(2, 11):
        silhouette_scores.append(   
            silhouette_score(Xm, MiniBatchKMeans(n_clusters = n_cluster).fit_predict(Xm))) 

    k = [2, 3, 4, 5, 6,7,8,9,10] 
    plt.figure(figsize=(12,8))
    plt.bar(k, silhouette_scores, color='red') 
    plt.title('Silhoutte Score Credit Card User(Mini Batch k-Means)',fontsize=20)
    plt.xticks(range(1,11),fontsize=11)
    plt.xlabel('Number of clusters', fontsize = 10) 
    plt.ylabel('Silhouette Score', fontsize = 10) 
plt.show()    
```

Dari bar plot metode silhouette untuk mini batch k-means di atas diperoleh sillhouette terbaik adalah 2 dan 4. 
Selanjutnya pada dataset ini kami akan menggunakan k-means clustering dan memilih 5 sebagai jumlah klaster.

```python
time = stop-start
print("waktu yang dibutuhkan adalah",time,"detik")
```

```python
start_mb = process_time()
MBKMeanModel = MiniBatchKMeans(n_clusters=5, init="k-means++", n_init=10, max_iter=300)
MBKMeanModel.fit(Xm)
labels_mb = MBKMeanModel.labels_
labels_mb
stop_mb = process_time()
```

```python
time_mb = stop_mb-start_mb
print("waktu yang dibutuhkan adalah",time_mb,"detik")
```

```python
#jaccard similarity
#jaccard similarity
def jaccard(list1, list2):
    intersection = 0
    for i in range(len(list1)):
        if list1[i]==list2[i]:
            intersection+=1
    union = (len(list1) + len(list2)) - intersection
    return float(intersection) / union

l = labels.tolist()
c = labels_mb.tolist()
jaccard(l,c)
```

Karena perbedaan time process kedua model tidak jauh berbeda, kami akan menggunakan k-means clustering.

<h2> Visualisasi Klaster Menggunakan K-Means </h2>

```python
#Transformasi data menjadi dua dimensi menggunakan PCA
dist = 1 - cosine_similarity(X)

pca = PCA(2)
pca.fit(dist)
X_PCA = pca.transform(dist)
X_PCA.shape
```

```python
X_PCA
```

```python
x, y = X_PCA[:, 0], X_PCA[:, 1]

colors = {0: 'red',
          1: 'blue',
          2: 'green', 
          3: 'yellow', 
          4: 'orange',  }

names = {0: 'Suku Membeli Barang MURAH dan DICICIL', 
         1: 'Paling JARANG Melakukan pembelian', 
         2: 'Pelanggan Konsumtif Membayar Tanpa DiCICIL', 
         3: 'Pelanggan Konsumtif Membayar dengan DICICIL', 
         4: 'Pelanggan KONSUMTIF Suka Membeli BARANG MAHAL dan DICICIL',
         }
  
df = pd.DataFrame({'x': x, 'y':y, 'label':labels}) 
groups = df.groupby('label')

fig, ax = plt.subplots(figsize=(20, 13)) 

for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5,
            color=colors[name],label=names[name], mec='none')
    ax.set_aspect('auto')
    ax.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off')
    ax.tick_params(axis= 'y',which='both',left='off',top='off',labelleft='off')
    
ax.legend(fontsize=15)
ax.set_title("Customers Segmentation Berdasarkan Aktivitas Pelanggan Kartu Kredit",fontsize=23)
plt.show()
```
